```python
def findtopviolations(df_name):
    rowcount = df_name['ViolationDescription'].count()
    unique_ticketdesc={}
    for iteration in range(rowcount):
        temp_val=df_name['ViolationDescription'].iloc[iteration]
        if(temp_val in unique_ticketdesc.keys()):
            unique_ticketdesc[temp_val]=unique_ticketdesc[temp_val]+1
        else:
            unique_ticketdesc[temp_val]=1

    #sorted(unique_ticketdesc, key=lambda kv: kv[1]['key3'], reverse=True)
    #for key, value in sorted(mydict.iteritems(), key=lambda (k,v): (v,k)):
    #    print "%s: %s" % (key, value)
    string="Top Violations:"
    space=" "
    index=1
    for w in sorted(unique_ticketdesc, key=unique_ticketdesc.get, reverse=True):
        #print (w, unique_ticketdesc[w])
        if (index<75):
            string=string+space+str(w)+space+str(unique_ticketdesc[w])
            index=index+1
    #print(string)
    return string
```
